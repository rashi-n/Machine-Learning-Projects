<h1> The .isin() Method

```python
df = pd.read_csv("Python Projects/Pandas Projects/pandas/employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
mask1 = df["Team"] == "Legal"
mask2 = df["Team"] == "Sales"
mask3 = df["Team"] == "Product"
df[mask1 | mask2|mask3]
```

```python
#USING .isin() Method

mask = df["Team"].isin(["Legal", "Sales", "Product"])
df[mask]
```

<h1> The .duplicated() Method

```python
df = pd.read_csv("Python Projects/Pandas Projects/pandas/employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name", inplace = True)
df.head(3)
```

```python
df[df["First Name"].duplicated()]
```

```python
df[df["First Name"].duplicated(keep = "last")]
df["First Name"].duplicated(keep = False) #This will True all duplicate rows
```

```python
# how to keep only UNIQUE ROWS, we can take opposite of above

~df["First Name"].duplicated(keep = False)
mask = ~df["First Name"].duplicated(keep = False)
df[mask]
```

<h1> The .drop_duplicates() Method

```python
df = pd.read_csv("Python Projects/Pandas Projects/pandas/employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name", inplace = True)
df.head(3)
```

```python
len(df)
```

```python
df.drop_duplicates()
```

```python
len(df.drop_duplicates())# count are still same even after drop duplicates because not any two rows are complete identical
```

```python
df.drop_duplicates(subset = ["First Name"], keep = "first")
df.drop_duplicates(subset = ["First Name"], keep = "last") #only last one is kept
df.drop_duplicates(subset = ["First Name"], keep = False)# it will not keep any row that has duplicates
```

```python
df.drop_duplicates(subset = ["First Name", "Team"])
```

```python
df.drop_duplicates(subset = ["First Name", "Team"], inplace = True)
df.head(2)
```

```python
len(df)
```
